```python
#genres
#id
# keywords
# overview
# title
# cast
# crew

movie = movie[['movie_id', 'title','genres','keywords','cast','crew','overview']]
```

```python
import ast
def convert (obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movie['genres'] =  movie['genres'].apply(convert)
```

```python
movie['keywords'] = movie['keywords'].apply(convert)
```

```python
movie.head()
```

```python
def convertCast (obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i['name'])
            counter += 1
        else:
            break
    return l
```

```python
movie['cast'] = movie['cast'].apply(convertCast)
```

```python
movie.head(1)
```

```python
def convertCrew (obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movie['crew'] = movie['crew'].apply(convertCrew)
```

```python
movie.head(1)
```

```python
movie['overview'] = movie['overview'].apply(lambda X:X.split())
```

```python
movie.head(1)
```

```python
movie['genres'] = movie['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movie['keywords'] = movie['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie['cast'] = movie['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie['crew'] = movie['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie.head(1)
```

```python
movie['tag'] = movie['genres'] + movie['keywords'] + movie['cast'] + movie['crew'] + movie['overview']
```

```python
movie.head(1)
```

```python
new_df = movie[['movie_id','title','tag']]
```

```python
new_df
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:" ".join(x))
```

```python
new_df['tag'] = new_df['tag'].apply(lambda x:x.lower())
```

```python
#ps(loving) ---> love
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    string = " ".join(y)
    return string
```

```python
new_df['tag'] = new_df['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vector = cv.fit_transform(new_df['tag']).toarray()
```

```python
vector
```

```python
# vector(4806 * 5000) , 4806 movies and each movie have a vector of size 5000
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity = cosine_similarity(vector)
```

```python
# similarity is a 4806 * 4806 matrix which contain cosine distance between one movie to all others
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title) 
```
